# Mass on Spring Simulations

VPython simulations of 1D spring-mass systems using Euler's method for numerical integration and their energy graphs.

## Systems covered
1. Single mass, undamped, undriven
2. Two coupled masses, undamped
3. Single mass, damped
4. Single mass, damped and driven 

Initial Stretch: 6 m

**Parameters:** `k = 1 N/m`, `m = 1 kg`, `dt = 0.001 s`, `l_0 = 10 m`

## 1. Single Mass — Undamped

```python
import numpy as np
from vpython import *
scene=canvas()
```

```python
# 1 mass, undamped, undriven

scene.width = 800
scene.center=vector(0, 2.5, 0)

ground = box(color=vector(0.5, 0.5, 0.5))
ground.pos = vec(0, -0.25, 0)
ground.size = vec(20, 0.5, 6)

xwall = 1
wall = box(color=color.red, texture=textures.rough)
wall.pos = vector(-ground.size.x/2 + xwall/2, 2.5, 0)
wall.size = vector(xwall, 5, 6)

R = 1       # m
m = 1       # kg
k = 1       # N/m
spring_l = 10  # m
coil_R = 0.3  # m

spring_x0 = wall.pos.x + xwall/2
spring = helix(color=color.white, radius=coil_R, thickness=coil_R/5)
spring.pos = vec(spring_x0, R, 0)   
spring.axis = vec(spring_l, 0, 0)

ball = sphere(pos=vector(spring.pos.x + spring.axis.x + R, R, 0), radius=R, color=color.blue)
ball.mass = m
ball.velocity = vector(0, 0, 0)

t = 0
dt = 1e-3

initial_stretch = 6 # m
if initial_stretch >= spring_l:
    initial_stretch = spring_l

ball.pos.x += initial_stretch
spring.axis.x += initial_stretch

graph(xtitle='Time (s)', ytitle='Energy (J)', foreground=color.black, background=color.white)

energy = gcurve(label='Total Energy')
kinetic = gcurve(color=color.red, label='Kinetic Energy')
potential = gcurve(color=color.blue, label= 'Potential Energy')

while t < 20:
    disp = spring.axis.x - spring_l
    F = - k * disp
    ball.acceleration = vec(F / m, 0, 0)
    ball.velocity += ball.acceleration * dt
    ball.pos += ball.velocity * dt

    spring.axis = vec(ball.pos.x - R - spring_x0, 0, 0)

    K = 0.5 * m * dot(ball.velocity, ball.velocity)
    U = 0.5 * k * disp**2
    E = K + U
    
    kinetic.plot(pos=(t, K))
    potential.plot(pos=(t, U))
    energy.plot(pos=(t, E))

    t += dt
    rate(1/dt)
    
    
```

## 2. Two Coupled Masses — Undamped

```python
import numpy as np
from vpython import *
scene2=canvas()
```

```python
# 2 masses, undamped, undriven

scene2.width = 800
scene2.center = vector(0, 2.5, 0)
scene2.range = 15

ground = box(color=vector(0.5, 0.5, 0.5))
ground.pos = vec(0, -0.25, 0)
ground.size = vec(40, 0.5, 6)

xwall = 1
wall = box(color=color.red, texture=textures.rough)
wall.pos = vector(-ground.size.x/2 + xwall/2, 2.5, 0)
wall.size = vector(xwall, 5, 6)

R = 1 # m

m1 = 1 # kg
m2 = 2 # kg

k = 1 # N/m
spring_l = 10 # m
coil_R = 0.3 # m

spring1_x0 = wall.pos.x + xwall/2
spring1 = helix(color=color.white, radius=coil_R, thickness=coil_R/5)
spring1.pos = vec(spring1_x0, R, 0)   
spring1.axis = vec(spring_l, 0, 0)

mass1 = sphere(pos=vector(spring1.pos.x + spring1.axis.x + R, R, 0), radius=R, color=color.blue)
mass1.mass = m1
mass1.velocity = vector(0, 0, 0)

spring2_x0 = mass1.pos.x + R
spring2 = helix(color=color.white, radius=coil_R, thickness=coil_R/5)
spring2.pos = vec(spring2_x0, R, 0)   
spring2.axis = vec(spring_l, 0, 0)

mass2 = sphere(pos=vector(spring2.pos.x + spring2.axis.x + R, R, 0), radius=R, color=color.blue)
mass2.mass = m2
mass2.velocity = vec(0, 0, 0)
t = 0
dt = 1e-3

initial_stretch = 6 # m
if initial_stretch >= spring_l:
    initial_stretch = spring_l 

mass1.pos.x += initial_stretch
mass2.pos.x += initial_stretch
spring1.axis.x += initial_stretch
spring2.axis.x += initial_stretch

tmax = 50

# Graph 1: total quantities 
gd1 = graph(align='left', width=400, height=300,
            xtitle='Time (s)', ytitle='Energy (J)', title='Total energy')
energy   = gcurve(graph=gd1, label='E total')
kinetic = gcurve(graph=gd1, color=color.red,   label='KE total')
potential     = gcurve(graph=gd1, color=color.blue,  label='PE total')

# Graph 2: per-component breakdown 
gd2 = graph(align='right', width=400, height=300,
            xtitle='Time (s)', ytitle='Energy (J)', title='Energy transfer')
kinetic1  = gcurve(graph=gd2, color=color.red,    label='KE1')
kinetic2  = gcurve(graph=gd2, color=color.orange, label='KE2')
potential1  = gcurve(graph=gd2, color=color.blue,   label='PE1')
potential2  = gcurve(graph=gd2, color=color.cyan,   label='PE2')

while t < tmax:
    rate(1/dt)
    
    disp1 = spring1.axis.x - spring_l
    disp2 = spring2.axis.x - spring_l
    F1 = - k * disp1
    F2 = - k * disp2
    mass1.acceleration = vec((F1 - F2)/ m1, 0, 0)
    mass1.velocity += mass1.acceleration * dt
    mass1.pos += mass1.velocity * dt

    mass2.acceleration = vec((F2)/ m2, 0, 0)
    mass2.velocity += mass2.acceleration * dt
    mass2.pos += mass2.velocity * dt

    spring1.axis = vec(mass1.pos.x - R - spring1_x0, 0, 0)
    
    spring2_x0 = mass1.pos.x + R
    spring2.pos.x = mass1.pos.x + R
    spring2.axis = vec(mass2.pos.x - R - spring2_x0, 0, 0)

    K1 = 0.5 * m1 * dot(mass1.velocity, mass1.velocity)
    K2 = 0.5 * m2 * dot(mass2.velocity, mass2.velocity)
    U1 = 0.5 * k * disp1**2   # spring 1: between wall and mass1
    U2 = 0.5 * k * disp2**2   # spring 2: between mass1 and mass2
    K = K1 + K2
    U = U1 + U2
    E = K + U

    kinetic.plot(pos=(t, K))
    potential.plot(pos=(t, U))
    energy.plot(pos=(t, E))

    kinetic1.plot(pos=(t, K1))
    kinetic2.plot(pos=(t, K2))
    potential1.plot(pos=(t, U1))
    potential2.plot(pos=(t, U2))
    

    t += dt
    
```

## 3. Single Mass — Damped

Linear drag coefficient $b = 0.1$ (underdamped)

```python
import numpy as np
from vpython import *
scene3=canvas()
```

```python
# 1 mass, damped, undriven

scene3.width = 800
scene3.center=vector(0, 2.5, 0)

ground = box(color=vector(0.5, 0.5, 0.5))
ground.pos = vec(0, -0.25, 0)
ground.size = vec(20, 0.5, 6)

xwall = 1
wall = box(color=color.red, texture=textures.rough)
wall.pos = vector(-ground.size.x/2 + xwall/2, 2.5, 0)
wall.size = vector(xwall, 5, 6)

R = 1       # m

m = 1       # kg

k = 1       # N/m
spring_l = 10  # m
coil_R = 0.3  # m

spring_x0 = wall.pos.x + xwall/2
spring = helix(color=color.white, radius=coil_R, thickness=coil_R/5)
spring.pos = vec(spring_x0, R, 0)   
spring.axis = vec(spring_l, 0, 0)

ball = sphere(pos=vector(spring.pos.x + spring.axis.x + R, R, 0), radius=R, color=color.blue)
ball.mass = m
ball.velocity = vector(1e-6, 0, 0)

initial_stretch = vec(6, 0, 0) # m
if mag(initial_stretch) >= spring_l:
    initial_stretch = norm(initial_stretch) * (spring_l - R)

ball.pos += initial_stretch
spring.axis += initial_stretch

# damping
damper = box(color=color.yellow, opacity=0.5)
fluid_start = spring_x0
fluid_end = ground.pos.x + ground.size.x / 2
damper.pos = vec((fluid_start + fluid_end) / 2, 2.5, 0)
damper.size = vec(fluid_end - fluid_start, wall.size.y, 6)

omega = np.sqrt(k/m)
b = 0.1

t = 0
dt = 1e-3

graph(xtitle='Time (s)', ytitle='Energy (J)', foreground=color.black, background=color.white)

energy = gcurve(label='Total Energy')
kinetic = gcurve(color=color.red, label='Kinetic Energy')
potential = gcurve(color=color.blue, label= 'Potential Energy')

while ball.velocity.mag > 1e-7:
    rate(1/dt)

    spring.pos.x = spring_x0

    disp = spring.axis - spring_l * norm(spring.axis)
    Fs = -k * disp

    Fdamp = - b * ball.velocity

    ball.acceleration = (Fs + Fdamp) / m
    ball.velocity += ball.acceleration * dt
    ball.pos += ball.velocity * dt

    spring.axis = vec(ball.pos.x - R - spring_x0, 0, 0)

    K = 0.5 * m * dot(ball.velocity, ball.velocity)
    U = 0.5 * k * mag(disp)**2
    E = K + U
    
    kinetic.plot(pos=(t, K))
    potential.plot(pos=(t, U))
    energy.plot(pos=(t, E))

    t += dt
    
```

## 4. Single Mass — Damped and Driven

Linear drag coefficient $b = 0.1$

Driving Amplitude $A_f = 0.5$

Driving frequency $\omega = \omega_0$ (resonance)

```python
import numpy as np
from vpython import *
scene4 = canvas(width=500, height=400, align='left')
```

```python
# 1 mass, damped, driven
scene4.center=vector(0, 2.5, 0)
scene4.range = 10

scene4.autoscale=False

ground = box(color=vector(0.5, 0.5, 0.5))
ground.pos = vec(0, -0.25, 0)
ground.size = vec(20, 0.5, 6)

# driver parameters
Af = 0.5
xwall = 1

wall = box(color=color.red, texture=textures.rough)
wall.pos = vector(-ground.size.x/2 + xwall/2, 2.5, 0)
wall.size = vector(xwall, 5, 6)

R = 1       # m
m = 1       # kg
k = 1       # N/m
omega_0 = np.sqrt(k / m)
spring_l = 10  # natural length, m
coil_R = 0.3  # m

spring_x0 = wall.pos.x + xwall/2
spring = helix(color=color.white, radius=coil_R, thickness=coil_R/5)
spring.pos = vec(spring_x0, R, 0)
spring.axis = vec(spring_l, 0, 0)

# Place ball at natural end of spring
ball = sphere(radius=R, color=color.blue)
ball.pos = vector(spring.pos.x + spring.axis.x + R, R, 0)
ball.mass = m
ball.velocity = vector(0, 0, 0)

omega_drive = omega_0 / 2

# damping
damper = box(color=color.yellow, opacity=0.5)
fluid_start = spring_x0
fluid_end = ground.pos.x + ground.size.x / 2
damper.pos = vec((fluid_start + fluid_end) / 2, 2.5, 0)
damper.size = vec(fluid_end - fluid_start, wall.size.y, 6)

b = 0.1

# Apply initial stretch
initial_stretch = vec(6, 0, 0)  # m  
if mag(initial_stretch) >= spring_l:
    initial_stretch = norm(initial_stretch) * (spring_l - 0.01)


spring.axis += initial_stretch
ball.pos = vec(spring.pos.x + spring.axis.x + R, R, 0)

ra0 = wall.pos + vec(0, 0, 0)

t = 0
dt = 0.001

graph(width=400, height=400, align='right', xtitle='Time (s)', ytitle='Energy (J)', foreground=color.black, background=color.white)

energy = gcurve(label='Total Energy')
kinetic = gcurve(color=color.red, label='Kinetic Energy')
potential = gcurve(color=color.blue, label= 'Potential Energy')

while t < 50:
    rate(1/dt) 

    wall.pos = vec(Af * np.sin(omega_drive * t) + ra0.x, ra0.y, ra0.z)  

    # Spring force
    spring.pos.x = wall.pos.x + xwall / 2
    L = ball.pos.x - spring.pos.x - R          
    disp = L - spring_l
    Fs = vec(-k * disp, 0, 0) 

    # Damping
    Fdamp = - b * ball.velocity

    ball.acceleration = (Fs + Fdamp) / m

    ball.velocity += ball.acceleration * dt
    ball.pos += ball.velocity * dt

    spring.axis = vec(ball.pos.x - spring.pos.x - R, 0, 0)

    K = 0.5 * m * dot(ball.velocity, ball.velocity)
    U = 0.5 * k * disp**2
    E = K + U
    
    kinetic.plot(pos=(t, K))
    potential.plot(pos=(t, U))
    energy.plot(pos=(t, E))

    t += dt
```
