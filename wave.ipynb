# Waves

## 1D Standing Wave - Fixed Ends

```python
import numpy as np
from vpython import *
scene1=canvas()
```

```python
# Left and Right Walls
wall1 = box(pos=vec(-1, 0, 0), length=0.01, width=1, height=1) # left
wall2 = box(pos=vec(1, 0, 0), length=0.01, width=1, height=1) # right

# String
N = 200
R = 0.01
balls = []

# Adding spheres to make string
for i in range(N):
    balls.append(sphere(color=color.blue, radius=R))

# Setting sphere positions
x = np.linspace(wall1.pos.x + R, wall2.pos.x - R, num=N)
for i in range(N):
    balls[i].pos.x = x[i]

# Parameter for wave equation d^2/dt^2(y) - c^2 d^2/dx^2(y) = 0
c = 1 # m/s

L = x[-1] - x[0]
x_local = x - x[0]

# Initial conditions
def g(x): # y(x, 0)
    return 0.5 * np.exp(-200 * (x- L/2)**2)
def h(x): # y_t(x, 0)
    return np.zeros_like(x)

# Calculating sine Fourier expansion coefficient
N_coeffs = 50
A = np.zeros(N_coeffs)
for n in range(1, N_coeffs + 1):
    integrand = g(x_local) * np.sin(n * np.pi * x_local / L)
    A[n-1] = (2 / L) * np.trapz(integrand, x_local)

    
# Time steps
t = 0
dt = 1e-3

while t < 10:                               
    rate(1/dt)
    y = sum(A[n-1] * np.sin(n * np.pi * x_local / L) * np.cos(n * np.pi * c * t / L)
        for n in range(1, N_coeffs + 1))
    
    for i, ball in enumerate(balls):
        ball.pos = vec(x[i], y[i], 0)

    t += dt

```

## Standing Wave - Free Ends

```python
import numpy as np
from vpython import *
scene2=canvas()
```

```python
# String
N = 200
R = 0.01
balls = []

# Adding spheres to make string
for i in range(N):
    balls.append(sphere(color=color.blue, radius=R))

# Setting sphere positions
x = np.linspace(-1 + R, 1 - R, num=N)
for i in range(N):
    balls[i].pos.x = x[i]

# Parameter for wave equation d^2/dt^2(y) - c^2 d^2/dx^2(y) = 0
c = 1 # m/s

L = x[-1] - x[0]
x_local = x - x[0]

# Initial conditions
def g(x): # y(x, 0)
    return 0.5 * np.exp(-200 * x**2)
def h(x): # y_t(x, 0)
    return np.zeros_like(x)

# Calculating cosine Fourier expansion coefficient
N_coeffs = 50
A = np.zeros(N_coeffs)
A[0] = 1/L * np.trapz(g(x_local), x_local)
for n in range(N_coeffs):
    integrand = g(x_local) * np.cos(n * np.pi * x_local / L)
    A[n] = (2 / L) * np.trapz(integrand, x_local)

    
# Time steps
t = 0
dt = 1e-3

while t < 10:                               
    rate(1/dt)
    y = sum(A[n] * np.cos(n * np.pi * x_local / L) * np.cos(n * np.pi * c * t / L)
        for n in range(N_coeffs))
    
    for i, ball in enumerate(balls):
        ball.pos = vec(x[i], y[i], 0)

    t += dt
```

## Standing Wave - One free end

```python
import numpy as np
from vpython import *
scene3=canvas()
```

```python
# Left Wall
wall1 = box(pos=vec(-1, 0, 0), length=0.01, width=1, height=1) # left

# String
N = 200
R = 0.01
balls = []

# Adding spheres to make string
for i in range(N):
    balls.append(sphere(color=color.blue, radius=R))

# Setting sphere positions
x = np.linspace(wall1.pos.x + R, -wall1.pos.x - R, num=N)
for i in range(N):
    balls[i].pos.x = x[i]

# Parameter for wave equation d^2/dt^2(y) - c^2 d^2/dx^2(y) = 0
c = 1 # m/s

L = x[-1] - x[0]
x_local = x - x[0]

# Initial conditions
def g(x): # y(x, 0)
    return 0.5 * np.exp(-200 * (x - L/2)**2)
def h(x): # y_t(x, 0)
    return np.zeros_like(x)

# Calculating Fourier expansion coefficient
N_coeffs = 50
A = np.zeros(N_coeffs)
for n in range(1, N_coeffs + 1):
    integrand = g(x_local) * np.sin((2*n - 1) * np.pi * x_local / (2*L))
    A[n-1] = (2 / L) * np.trapz(integrand, x_local)

    
# Time steps
t = 0
dt = 1e-3

while t < 10:                               
    rate(1/dt)
    y = sum(A[n-1] * np.sin((2*n - 1) * np.pi * x_local / (2 * L)) *  np.cos((2*n-1) * np.pi * c * t / (2 * L))
        for n in range(1, N_coeffs + 1))
    
    for i, ball in enumerate(balls):
        ball.pos = vec(x[i], y[i], 0)

    t += dt
```
