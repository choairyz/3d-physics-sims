# Pendulum Simulations

## Parameters
$m = 1$ kg

## 1. Simple pendulum, small angle

$\theta_0 = 6\degree$

```python
from vpython import *
import numpy as np
scene = canvas()
```

```python
scene.center=vec(0, 0.5, 0)

theta_i = np.deg2rad(6)
rod_l = 1
m = 1
g = vec(0, -9.81, 0)

ceil = box(pos=vec(0, rod_l, 0), length=1, height=0.01, width=0.5)

mass = sphere(pos=vec(0, rod_l, 0) + vec(np.sin(theta_i) * rod_l, -np.cos(theta_i) * rod_l, 0), radius=0.05,
              color=color.blue, make_trail=False, retain=20)

rod = cylinder(pos=vec(0, rod_l, 0), axis=vec(0, -1, 0), radius=0.01, length=rod_l, color=color.red)

omega = 0
curr_angle = theta_i

graph(xtitle='Time (s)', ytitle='Energy (J)', title='Mass Energy')

energy = gcurve()
kinetic = gcurve(color=color.red, label='Kinetic Energy')
potential = gcurve(color=color.blue, label='Potential Energy')

t = 0
dt = 1e-3

while t < 5 * (2 * np.pi * np.sqrt(rod_l / mag(g))):
    alpha = -(mag(g) / rod_l) * np.sin(curr_angle)
    curr_angle += omega * dt + 0.5 * alpha * dt**2
    alpha_new = -(mag(g) / rod_l) * np.sin(curr_angle)
    omega += 0.5 * (alpha + alpha_new) * dt

    mass.pos = vec(0, rod_l, 0) + vec(np.sin(curr_angle), -np.cos(curr_angle), 0) * rod_l
    rod.axis = mass.pos - vec(0, rod_l, 0)

    mass.velocity = vec(rod_l * omega * np.cos(curr_angle), rod_l * omega * np.sin(curr_angle), 0)

    K = 0.5 * m * dot(mass.velocity, mass.velocity)
    U = m * mag(g) * mass.pos.y
    E = K + U

    energy.plot(pos=(t, E))
    kinetic.plot(pos=(t, K))
    potential.plot(pos=(t, U))

    rate(1/dt)
    t += dt
```
